```python
def abc(n):
    a=0
    i=0
    while i<len(n):
        if n[i]==13 :
            i+=2
            continue
        a+=n[i]
        i=i+1
    return a  
n=[1,2,13,4,5,1] 
print("sum : ",abc(n))
```

```python
n=input("enter a integer : ")

def is_v(n):
    stack=[]
    for i in n:
        if i in "[({":
            stack.append(i)
        else:
            if not stack:
                return False
            top=stack.pop()    
            if i=="]" and top!="[" :
                 return False
            elif i=="}" and top!="{" :
                 return False
            elif i==")" and top!="(" :
                 return False
    return len(stack)==0
if is_v(n):
    print("valid")
else:
    print("invalid")
            
```
