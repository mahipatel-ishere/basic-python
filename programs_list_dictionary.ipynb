```python
n=input("enter string :")
c=input("enter char to count :")
f=input("enter char to fnd :")

def countchar(n,c):
    co=0
    for i in n:
        if i==c:
            co+=1
    return co 
def findchar(n,f):
    for i in range(len(n)):
        if n[i]==f:
             return 1 
    return -1        
r=countchar(n,c) 
s=findchar(n,f) 
print(r)
print(s)
```
