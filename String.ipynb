#slicing

```python
s="Hello How Are You"
print(s[0:-1])
print(s[0:6])
print(s[14:])
print(s[:8])
print(s[0:])
print(s[::-1])
print(s[-1:-3:-1])
print(s[-3:3:-1])
print(s[-3:3:-3])
print(s[1::-1])
print(s[3::-1])
print(s[3::-1])
print(s[-5::])
print(s[:-1:3])
```

```python
s="Hello How Are You"
index=0
for i in s:
    print(f"s[{index}]:{i}")
    index+=1
```
